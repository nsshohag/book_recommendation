## popularity based recommendation

```python
ratings_with_name=ratings.merge(books,on='ISBN')
ratings_with_name
```

```python
ratings_with_name.groupby('Book-Title').count()
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
all_books=num_rating_df.merge(avg_rating_df,on='Book-Title')
all_books
```

```python
all_books=all_books[all_books['num_rating']>0].sort_values('avg_rating',ascending=False)
popular_df=popular_df[popular_df['num_rating']>250].sort_values('avg_rating',ascending=False).head(50)
popular_df
all_books
```

```python
all_books=all_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_rating','avg_rating','Image-URL-M','ISBN']]
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_rating','avg_rating','Image-URL-M','ISBN']]
popular_df
all_books
c= ['Animal Farm']
```

## Collaborative Filtering Based Recommender System

```python
z=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
good_users=z[z].index
good_users.shape
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(good_users)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
similarity_scores[0]
```

```python
similarity_scores[0].shape
```

```python
def recommend(book_name):
    # index fetch
    index=np.where(pt.index==book_name)[0][0]
    print(index)
    distances= similarity_scores[index]
    similar_items=sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item=[]
        print(i[0])
        print(pt.index[i[0]])
        
        # to have all the information of the book
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        temp_df=temp_df.drop_duplicates('Book-Title')
        
        #item.append(list(temp_df['Book-Title'].values))
        
        # append bad diye extend korchi zate 2d array na hoy list
        item.extend(list(temp_df['Book-Title'].values))
        item.extend(list(temp_df['Book-Author'].values))
        item.extend(list(temp_df['Image-URL-M'].values))
        
        data.append(item)
        
    return data
# np.where(pt.index=='4 Blondes')[0][0]
recommend("Angels &amp; Demons")
```
